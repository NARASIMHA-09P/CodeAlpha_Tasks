```python
# TASK 4: SENTIMENT ANALYSIS
# Analyzing Amazon product reviews

print("="*60)
print("💬 TASK 4: SENTIMENT ANALYSIS")
print("="*60)

# ============================================
# STEP 1: Install and Import Tools
# ============================================

print("\n📦 Step 1: Installing TextBlob...")
!pip install textblob -q

from textblob import TextBlob
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import files

print("✅ Tools installed and loaded!")

# ============================================
# STEP 2: Create Sample Reviews (Amazon products)
# ============================================

print("\n📝 Step 2: Loading Amazon product reviews...")

# Here are 20 sample Amazon reviews for different products
reviews_data = {
    'Review': [
        # Positive reviews (10)
        "This product is amazing! Best purchase I ever made. Works perfectly.",
        "Absolutely love it! Fast shipping and great quality. Will buy again.",
        "Excellent product! exceeded my expectations. Highly recommend.",
        "Good value for money. Works as described. Happy with my purchase.",
        "Five stars! Very satisfied with this item. Great customer service.",
        "This is fantastic! Really helped me solve my problem. Worth every penny.",
        "Love it! Very easy to use and setup was simple. Great design.",
        "Perfect! Exactly what I needed. Arrived on time and in good condition.",
        "Awesome product! Very durable and well made. Would buy again.",
        "Great quality! Works like a charm. Very happy customer.",

        # Negative reviews (5)
        "Terrible product! Broke after 2 days. Waste of money.",
        "Very disappointed. Doesn't work as advertised. Poor quality.",
        "Worst purchase ever! Customer service was no help. Avoid this product.",
        "Cheap material and broke easily. Not worth the price. Very unhappy.",
        "Defective item received. Return process was difficult. Very frustrating.",

        # Neutral reviews (5)
        "It's okay. Does the job but nothing special.",
        "Average product. Works but has some issues. Fine for the price.",
        "Not bad, not great. Gets the job done. Might buy again.",
        "Decent product. Some good features, some bad. It's fine.",
        "Mixed feelings. Works sometimes but has glitches. It's acceptable."
    ]
}

# Convert to DataFrame
df = pd.DataFrame(reviews_data)

print(f"✅ Loaded {len(df)} product reviews")
print("\n📋 Sample reviews:")
print("-"*50)
for i in range(3):
    print(f"Review {i+1}: {df['Review'][i][:60]}...")
    print()

# ============================================
# STEP 3: Analyze Sentiment of Each Review
# ============================================

print("\n🔍 Step 3: Analyzing sentiment of each review...")
print("="*50)

# Function to get sentiment
def get_sentiment(text):
    blob = TextBlob(text)
    polarity = blob.sentiment.polarity  # -1 (negative) to +1 (positive)

    if polarity > 0.1:
        return 'Positive', polarity
    elif polarity < -0.1:
        return 'Negative', polarity
    else:
        return 'Neutral', polarity

# Analyze each review
sentiments = []
polarities = []

for review in df['Review']:
    sentiment, polarity = get_sentiment(review)
    sentiments.append(sentiment)
    polarities.append(polarity)

# Add results to DataFrame
df['Sentiment'] = sentiments
df['Polarity'] = polarities

print("\n✅ Analysis complete!")
print("\n📊 Results:")
print(df[['Review', 'Sentiment']].head(10))

# ============================================
# STEP 4: Show Detailed Results
# ============================================

print("\n" + "="*50)
print("📊 STEP 4: SENTIMENT RESULTS")
print("="*50)

# Count how many of each sentiment
sentiment_counts = df['Sentiment'].value_counts()
print("\n📈 Sentiment Distribution:")
print(sentiment_counts)

# Show examples of each sentiment
print("\n" + "="*50)
print("👍 POSITIVE REVIEW EXAMPLE:")
print("-"*50)
positive_example = df[df['Sentiment'] == 'Positive']['Review'].iloc[0]
print(f"Review: {positive_example}")
print(f"Sentiment Score: {df[df['Sentiment'] == 'Positive']['Polarity'].iloc[0]:.2f}")

print("\n" + "="*50)
print("👎 NEGATIVE REVIEW EXAMPLE:")
print("-"*50)
negative_example = df[df['Sentiment'] == 'Negative']['Review'].iloc[0]
print(f"Review: {negative_example}")
print(f"Sentiment Score: {df[df['Sentiment'] == 'Negative']['Polarity'].iloc[0]:.2f}")

print("\n" + "="*50)
print("😐 NEUTRAL REVIEW EXAMPLE:")
print("-"*50)
neutral_example = df[df['Sentiment'] == 'Neutral']['Review'].iloc[0]
print(f"Review: {neutral_example}")
print(f"Sentiment Score: {df[df['Sentiment'] == 'Neutral']['Polarity'].iloc[0]:.2f}")

# ============================================
# STEP 5: Create Visualization Charts
# ============================================

print("\n📊 Step 5: Creating charts...")

# Chart 1: Bar chart of sentiment counts
plt.figure(figsize=(8,5))
colors = ['green', 'red', 'gray']
bars = plt.bar(sentiment_counts.index, sentiment_counts.values, color=colors, edgecolor='black')

# Add numbers on top of bars
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height + 0.1,
             f'{int(height)}', ha='center', va='bottom', fontsize=12, fontweight='bold')

plt.title('SENTIMENT ANALYSIS RESULTS - Amazon Reviews', fontsize=14, fontweight='bold')
plt.xlabel('Sentiment', fontsize=12)
plt.ylabel('Number of Reviews', fontsize=12)
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.savefig('sentiment_barchart.png', dpi=200)
plt.show()
print("✅ Chart 1 saved: sentiment_barchart.png")

# Chart 2: Pie chart
plt.figure(figsize=(8,8))
plt.pie(sentiment_counts.values, labels=sentiment_counts.index,
        autopct='%1.1f%%', colors=['green', 'red', 'gray'],
        explode=(0.05, 0.05, 0), shadow=True)
plt.title('SENTIMENT DISTRIBUTION', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('sentiment_pie.png', dpi=200)
plt.show()
print("✅ Chart 2 saved: sentiment_pie.png")

# ============================================
# STEP 6: Calculate Percentages
# ============================================

print("\n" + "="*50)
print("📊 STEP 6: PERCENTAGE BREAKDOWN")
print("="*50)

total = len(df)
positive_pct = (sentiment_counts.get('Positive', 0) / total) * 100
negative_pct = (sentiment_counts.get('Negative', 0) / total) * 100
neutral_pct = (sentiment_counts.get('Neutral', 0) / total) * 100

print(f"\n👍 Positive Reviews: {positive_pct:.1f}% ({sentiment_counts.get('Positive', 0)} reviews)")
print(f"👎 Negative Reviews: {negative_pct:.1f}% ({sentiment_counts.get('Negative', 0)} reviews)")
print(f"😐 Neutral Reviews: {neutral_pct:.1f}% ({sentiment_counts.get('Neutral', 0)} reviews)")

# ============================================
# STEP 7: Business Insights
# ============================================

print("\n" + "="*50)
print("💡 STEP 7: BUSINESS INSIGHTS")
print("="*50)

print("""
Based on the sentiment analysis:

✅ WHAT'S WORKING WELL:
• {:.0f}% of reviews are positive
• Customers like the product quality and ease of use
• Fast shipping and good customer service mentioned

❌ WHAT NEEDS IMPROVEMENT:
• {:.0f}% of reviews are negative
• Quality issues and defective products mentioned
• Return process needs improvement

📊 RECOMMENDATIONS:
1. Investigate quality control issues
2. Improve return process
3. Highlight positive features in marketing
4. Address common complaints in product updates
""".format(positive_pct, negative_pct))

# ============================================
# STEP 8: Save Results
# ============================================

print("\n💾 Step 8: Saving results...")

# Save DataFrame to CSV
df.to_csv('sentiment_analysis_results.csv', index=False)
print("✅ Results saved to: sentiment_analysis_results.csv")

# Download all files
print("\n📥 Downloading files...")
files.download('sentiment_analysis_results.csv')
files.download('sentiment_barchart.png')
files.download('sentiment_pie.png')

```
