```python
# TASK 1: WEB SCRAPING - TITANIC DATA

# STEP 1: Install needed tools (Google Colab does this automatically)
print("Step 1: Loading tools...")

# STEP 2: Import libraries
import pandas as pd
from google.colab import files

print("✅ Tools loaded successfully!")
print("="*50)

# STEP 3: Get Titanic data from web
print("\nStep 2: Loading Titanic data...")

# This is a direct link to Titanic dataset
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"

# Load the data
data = pd.read_csv(url)

print("✅ Data loaded successfully!")
print("="*50)

# STEP 4: Look at the data
print("\nStep 3: Previewing the data...")
print("\nFirst 5 rows:")
print(data.head())

print("\n" + "="*50)

# STEP 5: Basic information
print("\nStep 4: Dataset information...")
print(f"Total passengers: {len(data)}")
print(f"Column names: {list(data.columns)}")

print("\n" + "="*50)

# STEP 6: Save to file
print("\nStep 5: Saving to CSV file...")
data.to_csv('titanic_data.csv', index=False)
print("✅ Saved as 'titanic_data.csv'")

print("\n" + "="*50)

# STEP 7: Download to your computer
print("\nStep 6: Downloading file...")
files.download('titanic_data.csv')
print("✅ File downloaded to your computer!")

print("\n" + "="*50)
print("🎉 TASK 1 COMPLETED SUCCESSFULLY!")
print("="*50)

# STEP 8: Show summary
print("\n📊 QUICK SUMMARY OF YOUR DATA:")
print("-"*30)
print(f"📌 Total records: {len(data)}")
print(f"📌 Columns: 12")
print(f"📌 File saved: titanic_data.csv")
print(f"📌 Location: Your Downloads folder")

print("\n✅ You can now use this file for your analysis!")
```
