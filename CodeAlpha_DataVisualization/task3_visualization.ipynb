```python
# TASK 3: DATA VISUALIZATION 

import pandas as pd
import matplotlib.pyplot as plt
from google.colab import files

# Load data
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

# Clean data
df['Age'].fillna(df['Age'].median(), inplace=True)

print("✅ Data loaded! Creating charts...\n")

# ============================================
# CHART 1: Gender Survival
# ============================================
female = df[df['Sex']=='female']['Survived'].mean() * 100
male = df[df['Sex']=='male']['Survived'].mean() * 100

plt.figure(figsize=(6,4))
plt.bar(['Women', 'Men'], [female, male], color=['pink', 'lightblue'])
plt.title('Survival by Gender')
plt.ylabel('Survival (%)')
plt.ylim(0,100)
plt.savefig('chart1.png')
plt.show()
print("✅ Chart 1: gender_survival.png")

# ============================================
# CHART 2: Class Survival
# ============================================
class1 = df[df['Pclass']==1]['Survived'].mean() * 100
class2 = df[df['Pclass']==2]['Survived'].mean() * 100
class3 = df[df['Pclass']==3]['Survived'].mean() * 100

plt.figure(figsize=(6,4))
plt.bar(['1st Class', '2nd Class', '3rd Class'], [class1, class2, class3],
        color=['gold', 'silver', '#CD7F32'])
plt.title('Survival by Passenger Class')
plt.ylabel('Survival (%)')
plt.ylim(0,100)
plt.savefig('chart2.png')
plt.show()
print("✅ Chart 2: class_survival.png")

# ============================================
# CHART 3: Age Distribution
# ============================================
plt.figure(figsize=(8,4))
plt.hist(df[df['Survived']==0]['Age'], bins=20, alpha=0.6, label='Did Not Survive', color='red')
plt.hist(df[df['Survived']==1]['Age'], bins=20, alpha=0.6, label='Survived', color='green')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.legend()
plt.savefig('chart3.png')
plt.show()
print("✅ Chart 3: age_distribution.png")

# ============================================
# CHART 4: Overall Survival (Pie Chart)
# ============================================
survived = df['Survived'].sum()
died = len(df) - survived

plt.figure(figsize=(6,6))
plt.pie([survived, died], labels=['Survived', 'Did Not Survive'],
        autopct='%1.1f%%', colors=['lightgreen', 'lightcoral'])
plt.title('Overall Survival Rate')
plt.savefig('chart4.png')
plt.show()
print("✅ Chart 4: overall_survival.png")

# ============================================
# CHART 5: Fare by Class
# ============================================
plt.figure(figsize=(8,4))
df.boxplot(column='Fare', by='Pclass')
plt.title('Ticket Prices by Class')
plt.xlabel('Class (1=Rich, 3=Poor)')
plt.ylabel('Fare (£)')
plt.suptitle('')
plt.savefig('chart5.png')
plt.show()
print("✅ Chart 5: fare_by_class.png")

# ============================================
# SUMMARY & DOWNLOAD
# ============================================
print("\n" + "="*40)
print("🎉 5 CHARTS CREATED!")
print("="*40)
print("""
📊 WHAT YOU LEARNED:
• Women survived more than men
• Rich passengers survived more
• Children survived more than adults
• Only 38% of people survived
• Rich people paid higher fares

📁 Charts saved:
• chart1.png (Gender)
• chart2.png (Class)
• chart3.png (Age)
• chart4.png (Pie)
• chart5.png (Fare)
""")

```

```python
# Download all charts
for chart in ['chart1.png', 'chart2.png', 'chart3.png', 'chart4.png', 'chart5.png']:
    files.download(chart)

print("\n✅ All charts downloaded to your computer!")
print("🎉 TASK 3 COMPLETE!")
```
